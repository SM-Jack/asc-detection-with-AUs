```python
import csv
import os

import math
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_frames = []
for (root, dirs, file) in os.walk("../all_features"):
    for f in file:
        if ".csv" in f:
            path = root + "/" + f
            df = pd.read_csv(path)
            data_frames.append(df)
```

```python
example = data_frames[0]
print(example.head())
```

```python
# openface ranges from 0 to 5, so I will divide by 5
example.plot(kind='line', y=["openface_AU01_r", "libreface_AU01_i", "me_graph_AU1"], subplots=False, figsize=(10, 10))
plt.title('Intensity Over Time')
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.show()
```
